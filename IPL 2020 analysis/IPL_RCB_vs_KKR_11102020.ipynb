## How many no-ball will be bowled in the match?

```python
deli
```

```python
rcb_kkr = deli[(deli['bowling_team'] == 'Royal Challengers Bangalore')|(deli['bowling_team'] == 'Kolkata Knight Riders')]
rcb_kkr
```

```python
match_id_rk=rcb_kkr['match_id'].unique()
match_id_rk
```

```python
## total no balls in each rcb vs kkr match
no_balls = []
for i in match_id_rk:
    no_balls.append(rcb_kkr[rcb_kkr['match_id'] == i].noball_runs.sum())
no_balls
```

```python
no_ball_mean = statistics.mean(no_balls)
no_ball_mean
```

```python
no_ball_median = statistics.median(no_balls)
no_ball_median
```

```python
no_ball_mode = statistics.mode(no_balls)
no_ball_mode
```
