## 2.How many wickets will MI loose in overs 6-15 ? >>> 3-4

```python
mi = deli[deli['batting_team'] == 'Chennai Super Kings']
```

```python
mi
```

```python
mi_match=mi['match_id'].unique()
mi_match
```

```python
mi_n = mi[(mi.over >=6) & (mi.over <=15)]
```

```python
wickets_per_match = []
for i in mi_match:
    wickets_per_match.append(mi_n[mi_n['match_id'] == i].dismissal_kind.notna().sum())
wickets_per_match
```

```python
import statistics
```

```python
my_mean = statistics.mean(wickets_per_match)
my_mean
```

```python
my_median = statistics.median(wickets_per_match)
my_median
```

```python
my_mode = statistics.mode(wickets_per_match)
my_mode
```

```python
my_stdev = statistics.stdev(wickets_per_match)
my_stdev
```

```python
wickets_per_match.sort(reverse=False)
```

```python
sns.distplot(wickets_per_match)
```

```python
## As the data is normally distributed, and the frequency of data is highest in 2-4 region so mi will loose 3-4 wickets between overs 6-15
```

## 4.What will be the total run??

```python
## Today's match will be between KKR and MI
```

```python
total_match = deli['match_id'].unique()
total_match
```

```python
total_runs = []
for i in total_match:
    total_runs.append(deli[deli['match_id'] == i].total_runs.sum())
total_runs
```

```python
total_run_mean = statistics.mean(total_runs)
total_run_mean
```

```python
total_run_median = statistics.median(total_runs)
total_run_median
```

```python
total_run_mode = statistics.mode(total_runs)
total_run_mode
```

```python
total_run_std = statistics.stdev(total_runs)
total_run_std
```

```python
total_runs.sort(reverse=False)
```

```python
sns.distplot(total_runs)
```

```python
sns.boxplot(total_runs)
```

```python
## Although total run median and mode is quite close to each other but standard deviation is very high. from teh distribution it is clear that total runs will be less than 320
```

## 3.How many 4s will be hit by kkr in teh match? >>

```python
deli
```

```python
total_ball = deli[(deli['batting_team'] == 'Kolkata Knight Riders')&(deli['bowling_team'] == 'Mumbai Indians')]
total_ball
```

```python
four_run_ball = total_ball[total_ball.batsman_runs == 4]
four_run_ball
```

```python
match_id_rk=total_ball['match_id'].unique()
match_id_rk
```

```python
## total no balls in each rcb vs kkr match
total_four_every_match = []
for i in match_id_rk:
    total_four_every_match.append(four_run_ball[four_run_ball['match_id'] == i].total_runs.count())
total_four_every_match
```

```python
four_run_mean = statistics.mean(total_four_every_match)
four_run_mean
```

```python
four_run_median = statistics.median(total_four_every_match)
four_run_median
```

```python
four_run_mode = statistics.mode(total_four_every_match)
four_run_mode
```

```python
sns.boxplot(total_four_every_match)
```

```python
total_four_every_match.sort(reverse=False)
```

```python
sns.distplot(total_four_every_match)
```

```python
### Mean is 13 and median, mode of 4's are at 14 and histogram shows the highest density around 13-15 range. So the option is 13 or more
```
